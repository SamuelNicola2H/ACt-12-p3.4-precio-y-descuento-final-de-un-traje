## Las variables
1. CT --> Costo del Traje
2. DE --> Descuento que se obtendra
3. PF --> Precio final del traje

## Formulas
## $DE = CT * 0.15$
## $DE = CT * 0.08$
## $PF = CT - DE$

## Diagrama de fliujo Lucidchart

```python
# para cuando es verdadero (Si)
# Ejemplo 3.4
# Algoritmo para determinar el descuento y el precio final de un traje
# Calculando el Descuento final usando if else
# datos de entrada DE = CT, DE, PF
# Si es verdadero: DE = CT * 0.15
# Si es falso: CT * 0.08
# Cálculo del Precio Final

CT = float(input("Ingrese el costo del traje: "))
if CT > 2500:
    print("Condición: VERDADERO (DE = CT * 0.15)")
    DE = CT * 0.15
else:
    print("Condición: FALSO ( DE = CT * 0.08)")
    DE = CT * 0.08
# Precio Final
PF = CT - DE
print("El precio final es: ", PF)
print("El descuento es: ", DE)

print("Fin del programa")
print("Programa realizado por Samuel Nicola NL = 40")
```
